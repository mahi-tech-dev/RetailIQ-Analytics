```python
# ==========================================
# RetailIQ Analytics
# Milestone 4 - Exploratory Data Analysis
# ==========================================

import pandas as pd

df = pd.read_csv("../data/processed/clean_superstore.csv")

print(df.shape)
```

```python
# KPI Summary

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = df["Order ID"].nunique()

print("Total Sales:", round(total_sales,2))
print("Total Profit:", round(total_profit,2))
print("Total Orders:", total_orders)
```

```python
# Average Metrics

average_sales = df["Sales"].mean()
average_profit = df["Profit"].mean()

print("Average Sales:", round(average_sales,2))
print("Average Profit:", round(average_profit,2))
```

```python
# Category-wise Sales

df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
# Category-wise Profit

df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
```

```python
# Top 10 Products

df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# Top 10 Customers

df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# Region-wise Sales

df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
```
