```python
# ==========================================
# RetailIQ Analytics
# Milestone 6 - Business Insights
# ==========================================

import pandas as pd

df = pd.read_csv("../data/processed/clean_superstore.csv")

print(df.shape)
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print("Highest Revenue Category:")
print(category_sales)
```

```python
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

print("Highest Profit Category:")
print(category_profit)
```

```python
region_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

print("Region-wise Profit:")
print(region_profit)
```

```python
top_customer = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

print(top_customer)
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

print(top_products)
```

```python
loss_products = df.groupby("Product Name")["Profit"].sum().sort_values().head(10)

print(loss_products)
```
